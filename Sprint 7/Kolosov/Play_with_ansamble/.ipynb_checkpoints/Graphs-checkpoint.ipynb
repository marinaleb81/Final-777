```python
from bokeh.plotting import figure, show

data_models = ['Random Forest', 'XGBoost', 'LightGBM', 'Logistic Regression', 'SVM', 
               'Voting Ensemble', 'Bagging', 'Gradient Boosting', 'Stacking', 'AdaBoost']
f1_scores = [0.987, 0.979, 0.981, 0.855, 0.892, 0.979, 0.984, 0.913, 0.991, 0.988]

# sorting the bars means sorting the range factors
sorted_models = sorted(data_models, key=lambda x: f1_scores[data_models.index(x)])

p = figure(x_range=sorted_models, y_range=(0.7,1), width=800, height=500, title="F1 scores for different models",
           toolbar_location=None, tools="")

p.vbar(x=data_models, top=f1_scores, width=0.9)

p.xgrid.grid_line_color = None

show(p)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

d = {
    'Model': ['Stacking','AdaBoost','Random Forest','Bagging','LightGBM','Voting Ensemble',
              'XGBoost','Gradient Boosting','SVM','Logistic Regression'],
    'F1-Score': [0.991,0.988,0.987,0.984,0.981,0.979,0.979,0.913,0.892,0.855]
}

sns.barplot(y='Model', x='F1-Score', data=d, orient='h')
plt.xlim(0.8, 1)
plt.show()
```
